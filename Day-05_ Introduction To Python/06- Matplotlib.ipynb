# Matplotlib

- Matplotlib is a python library used for Data  visualization

- ## Line Plot

```python
import numpy as np

import pandas as pd

from matplotlib import pyplot as plt
```

```python
x= np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y) 
plt.show()
```

```python
# ADDING TITLE AND LABELS
```

```python
plt.plot(x,y)
plt.title('Line plot')
plt.xlabel('x-label')
plt.ylabel('y-label')          
plt.show()
```

```python
# CHANGING LINE AESTHETICS
```

```python
plt.plot(x,y,color="g", linestyle=':', linewidth=5)
plt.show()
```

## Adding two lines in the same plot 

```python
x=np.arange(1,11)
y1=2*x
y2=3*y
```

```python
plt.plot(x,y1,color="g", linestyle=':', linewidth=2)
plt.plot(x,y2,color="y", linestyle='-.', linewidth=3)
plt.title('Line plot')
plt.xlabel('x-label')
plt.ylabel('y-label') 
plt.grid(True)
plt.show()
```

## Adding sub Plots

```python
x=np.arange(1,11)
y1=2*x
y2=3*y

plt.subplot(1,2,1)
plt.plot(x,y1,color="g", linestyle=':', linewidth=2)

plt.subplot(1,2,2)
plt.plot(x,y2,color="r", linestyle=':', linewidth=2)

plt.show()
```

## Bar Plot

```python
student= {"Ahsan":80,"Mohammad Ehsan" : 70 , "Ihsan": 60, "Ali Raza" :50}
```

```python
names=list (student.keys())
values=list(student.values())
```

```python
names , values
```

```python
plt.bar(names, values)
plt.show()
```

```python
# Adding title and Label Into the Bar Plot
```

```python
plt.bar(names, values , color="b")
plt.title('Bar plot')
plt.xlabel('Names of Students')
plt.ylabel('Marks of Students') 
plt.grid(True)
plt.show()
```

## Horizoltal Bar Plot

```python
plt.barh(names, values , color="y")
plt.title("Bar Plot")
plt.xlabel("Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

## Scatter Plot

```python
# craeting a basic Scatter plot
a=[10,20,30,40,50,60,70]
b=[12,5,22,32,13,32,22]
plt.scatter(a,b)
plt.show()
```

```python
# cahnging marks Aesthetics
```

```python
a=[10,20,30,40,50,60,70]
b=[5,10,15,20,25,30,35]
plt.scatter(a,b, marker="*",c="g", s=100)
plt.show()
```

## Adding two Marker in Plot 

```python
x=[10,20,30,40,50,60,70]
a=[5,10,15,20,25,30,35]
b=[3,9,12,15,18,21,24]
plt.scatter(x,a, marker="*",c="g", s=100)
plt.scatter(x,b, marker=".",c="g", s=200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,a, marker="*",c="g", s=100)

plt.subplot(1,2,2)
plt.scatter(x,a, marker=".",c="g", s=200)
plt.show()
```

## Box- Plot

```python
# Creating Data
a=[1,2,3,4,5,6,7,8,9]
b=[1,2,3,4,5,4,3,2,1]
c=[6,7,8,9,8,7,6,5,4]
data=list([a,b,c])
```

```python
# Making plot
plt.boxplot(data)
plt.show()
```

## Violin-Plot

```python
plt.violinplot(data, showmedians=True)
plt.show()
```

## Pie chart

```python
# Creating Data
name=["ahsan","Muhammad","Ehsan", "Ihsan"]
marks=[90,80,70,60]
```

```python
# Making plot

plt.pie(marks, labels=name)
plt.show()
```

```python
# Changing Aesthetics
```

```python
plt.pie(marks, labels=name,autopct="% 0.1f%%", colors=["yellow","pink","green","grey"])
plt.show()
```
